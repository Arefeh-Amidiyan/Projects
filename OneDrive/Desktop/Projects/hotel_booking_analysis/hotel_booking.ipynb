## Handling Missing Values
In this section, missing values in different columns are replaced as follows:

1. **Column `children`**: Replace missing values with the mode (most frequent value).
2. **Column `country`**: Replace missing values with `'Unknown'`.
3. **Columns `agent` and `company`**: Replace missing values with `-1`.


```python
children_mode = df['children'].mode()[0]
df['children'].fillna(children_mode, inplace=True)

df['country'].fillna('Unknown', inplace=True)

df['agent'].fillna(-1, inplace=True)
df['company'].fillna(-1, inplace=True)
```

## 2. Convert the dtypes to appropriate ones according to the values
### Convert Datatypes of columns

```python
df['children'] = df['children'].astype(int)  
df['agent'] = df['agent'].astype(int)        
df['company'] = df['company'].astype(int)    
df['adr'] = df['adr'].astype(float)          

df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])

categorical_columns = [
    'hotel', 'arrival_date_month', 'meal', 'country', 
    'market_segment', 'distribution_channel', 'reserved_room_type',
    'assigned_room_type', 'deposit_type', 'customer_type', 'reservation_status'
]
for col in categorical_columns:
    df[col] = df[col].astype('category')
```

## 9.3. Comparison of Lead Time for Cancelled and Not Cancelled Bookings

This density plot compares the **lead time** (number of days between booking and arrival) for bookings that were either cancelled or not cancelled. The purpose of this analysis is to observe any patterns or trends in how lead time impacts the likelihood of cancellation.

- **Red Curve**: Represents bookings that were cancelled.
- **Green Curve**: Represents bookings that were not cancelled.

#### Observations:
- The plot allows us to see the distribution of lead times for both cancelled and not cancelled bookings.
- Longer lead times appear to have a higher density of cancellations, suggesting that guests who book further in advance might be more likely to cancel.
- Shorter lead times are more densely associated with non-cancelled bookings, indicating a higher chance of guests following through with their reservations when they book closer to the arrival date.

This visualization can be insightful for hotels looking to understand booking patterns and manage potential risks associated with cancellations.


```python
cancelled_lead_time = df[df['is_canceled'] == 1]['lead_time']
not_cancelled_lead_time = df[df['is_canceled'] == 0]['lead_time']

plt.figure(figsize=(16, 8))
sns.kdeplot(cancelled_lead_time, color='red', label='Cancelled', shade=True)
sns.kdeplot(not_cancelled_lead_time, color='green', label='Not Cancelled', shade=True)

plt.title("Comparison of Lead Time for Cancelled and Not Cancelled Bookings", weight="bold", color="gray", fontsize=18, loc="left")
plt.xlabel("Lead Time", loc="left", weight="bold", fontsize=14, color="gray")
plt.ylabel("Density", loc="bottom", weight="bold", fontsize=14, color="gray")
plt.legend()
ax = plt.gca()
ax.spines["left"].set_color("gray")
ax.spines["bottom"].set_color("gray")
ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)
ax.tick_params(axis="x", colors="gray", labelsize="large")
ax.tick_params(axis="y", colors="gray", labelsize="large")
plt.show()
```
